```python
x=np.array([1000,2000,3000,4000])
y=np.array([10,20,30,40])
plt.xlabel("Area")
plt.ylabel("Price in Lakhs")
plt.title("Relationship between Area and Price",color='Hotpink')
plt.grid()
plt.plot(x,y,marker='o',ms=10,mec='g',mfc='r')
plt.show()
```

```python
x=np.array([1000,2000,3000,4000])
y=np.array([10,20,30,40])
plt.xlabel("Area",fontsize=17,color='green')
plt.ylabel("Price in Lakhs",fontsize=17,color='green')
plt.title("Relationship between Area and Price",color='Hotpink',fontsize=20)
plt.grid(color='yellow')
plt.plot(x,y,marker='o',ms=13,mec='g',mfc='r',linestyle='dotted',color='c')
plt.show()
```
